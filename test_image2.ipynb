```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox
from PIL import Image, ImageTk

class FirstPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)

        #add image
        lp_load = Image.open('sample1.jpg')
        lp_photo = ImageTk.PhotoImage(lp_load)
        lp_ilabel = tk.Label(self, image=lp_photo)
        lp_ilabel.image=lp_photo
        lp_ilabel.place(x=0,y=0)

        #boarder to hold username and password
        lp_border = tk.LabelFrame(self, text='Login', bg='blue', bd=10, font='Arial, 20')
        lp_border.pack(fill='both', expand='yes', padx=150, pady=150)

        #Username
        lp_uname = tk.Label(lp_border, text='Username', font=('Arial bold', 15), bg='blue')
        lp_uname.place(x=50, y=20)
        lp_entuname = tk.Entry(lp_border, width=30, font='Arial 7', bd=5)
        lp_entuname.place(x=180, y=20)

        # Password
        lp_pword = tk.Label(lp_border, text='Password', font=('Arial bold', 15), bg='blue')
        lp_pword.place(x=50, y=80)
        lp_entpword = tk.Entry(lp_border, width=30, font='Arial 7', bd=5, show='*')
        lp_entpword.place(x=180, y=80)

        #submit button function
        def verify():
            #code for try...54.20
            try:
                with open('user account.txt', 'r') as f:
                    info = f.readlines()
                    # ....code to manage wrong input(i = 0)
                    i = 0
                    for e in info:
                        u, p = e.split(',')
                        #strip() is to clear spaces...52.46
                        if u.strip() == lp_entuname.get() and p.strip() == lp_entpword.get():
                            controller.show_frame(SecondPage)
                            i = 1
                            break
                    if i==0:
                        messagebox.showinfo('error', 'wrong input')
            except:
                messagebox.showinfo('error', 'wrong input')


        # submit button
        fp_btn = tk.Button(lp_border, text='Submit', bg="red", font=('Arial bold', 10),
                           command=verify)
        fp_btn.place(x=200, y=120)

        # ====================================================Register Window==========================================================

        # register button function
        def register():
            window = tk.Tk()
            window.resizable(0,0) #to mke d window non resizable(i.e kept portable)..39.40
            window.configure(bg='red') #for background color
            window.title('Register')

            # register username
            rg_uname = tk.Label(window, text='Username:', font=('Arial', 15), bg='sky blue')
            rg_uname.place(x=10, y=10)
            rg_entuname = tk.Entry(window, width=30, font='Arial 7', bd=5)
            rg_entuname.place(x=200, y=10)

            # Rg Password
            rg_pword = tk.Label(window, text='Password:', font=('Arial', 15), bg='sky blue')
            rg_pword.place(x=10, y=60)
            rg_entpword = tk.Entry(window, width=30, font='Arial 7', bd=5, show='*') #45.40(show='*')
            rg_entpword.place(x=200, y=60)

            # Rg confirm Password
            rg_cpword = tk.Label(window, text='Confirm Password:', font=('Arial', 15), bg='sky blue')
            rg_cpword.place(x=10, y=110)
            rg_entcpword = tk.Entry(window, width=30, font='Arial 7', bd=5, show='*')
            rg_entcpword.place(x=200, y=110)

            # Rg submit button function
            def rg_submit():
                if rg_entuname.get()!='' and rg_entpword.get()!='' and rg_entcpword.get()!='':
                    if rg_entpword.get()==rg_entcpword.get():
                        with open('user account.txt', 'a') as f:
                            f.write(rg_entuname.get()+','+rg_entpword.get()+'\n')
                            messagebox.showinfo('Welcome','Registration Successful')

                    else:
                        messagebox.showinfo('error','password not same!!!')
                else:
                    messagebox.showinfo('error','please fill empty boxes!!!')

            # Rg submit button
            rg_btn = tk.Button(window, text='Submit', bg="#ffc22a", font=('Arial bold', 10), command=rg_submit)
            rg_btn.place(x=170, y=150)

            window.geometry('470x220')
            window.mainloop()

        # register button
        fp_btn = tk.Button(lp_border, text='Register', bg="red", font=('Arial bold', 10),
                           command=register)
        fp_btn.place(x=300, y=120)
#====================================================2nd Page==========================================================


class  SecondPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)

        #page background colour
        self.configure(bg='tomato')

        sp_lbl = tk.Label(self, text='SecondPage', font='Arial, 10')
        sp_lbl.place(x=230, y=230)

        self.name = StringVar
        sp_ent = tk.Entry(self, textvariable=self.name, font='Arial 7')
        sp_ent.place(x=400, y=320)

        # submit button for personal details
        sp_btn = tk.Button(self, text='Next Page', bg="red", font='Arial, 10',
                           command=lambda: controller.show_frame(ThirdPage))
        sp_btn.place(x=650, y=450)
#====================================================3rd Page==========================================================


class ThirdPage(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)

        tp_lbl = tk.Label(self, text='ThirdPage', font='Arial, 10')
        tp_lbl.place(x=230, y=230)

        self.name = StringVar
        sp_ent = tk.Entry(self, textvariable=self.name, font='Arial 7')
        sp_ent.place(x=400, y=320)

        # submit button for personal details
        tp_btn = tk.Button(self, text='Next Page', bg="red", font='Arial, 10',
                           command=lambda: controller.show_frame(FirstPage))
        tp_btn.place(x=650, y=450)

#====================================================Application==========================================================
class Application(tk.Tk):
    def __init__(self, *args, **kwargs):
        tk.Tk.__init__(self, *args, **kwargs)


        # create a window
        window = tk.Frame(self)
        window.pack()

        window.grid_rowconfigure(0, minsize=500)
        window.grid_columnconfigure(0, minsize=800)

        self.frames = {}
        for F in (FirstPage, SecondPage, ThirdPage):
            frame = F(window, self)
            self.frames[F] = frame
            frame.grid(row=0, column=0, sticky='nsew')

        self.show_frame(FirstPage)

    def show_frame(self, page):
        frame = self.frames[page]
        frame.tkraise()
        self.title('Application')



app = Application()
app.maxsize(800,500)#...40.00 to limit how maximized the window will be
app.mainloop()
```
